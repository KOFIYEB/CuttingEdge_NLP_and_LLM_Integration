# **LOADING PRETRAINED MODEL**

Before embarking on the intricate process of data preparation and fine-tuning, it was imperative to load and test the entire base model in its original form from Hugging Face. This unmodified version of the model provided a benchmark for its performance on our specific domain. Evaluating the base model's capabilities on our data without any modifications allowed us to understand its default behavior and effectiveness in handling domain-specific prompts.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

device = "cuda" # the device to load the model onto

model = AutoModelForCausalLM.from_pretrained("mistralai/Mistral-7B-Instruct-v0.1")
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.1")

```

# **TEST DRIVE MODEL**

```python
messages = [
    {"role": "user", "content": "Write me a blog post about this topic: fitness tips"},

]

encodeds = tokenizer.apply_chat_template(messages, return_tensors="pt")

model_inputs = encodeds.to(device)
model.to(device)

generated_ids = model.generate(model_inputs, max_new_tokens=1000, do_sample=True)
decoded = tokenizer.batch_decode(generated_ids)
print(decoded[0])
```
